# Polynomial Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values # Modified to take data from Index 1 (second column)
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```
# Copied from simple_linear_regression.ipynb
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
# Since the dataset is not split into training and test sets,
# the fit should accomodate X and y dataset instead
lin_reg.fit(X, y)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures

# Normal Polynomial Regression Training (Commented Out)
# poly_reg = PolynomialFeatures(degree = 2)

# Altered Polynomial Regression Training (Degree = 4)
poly_reg = PolynomialFeatures(degree = 4)

X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression() # Second Linear Regression created
lin_reg_2.fit(X_poly, y)
```

## Visualising the Linear Regression results

```
# Visualizing the Linear Regression Data
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg.predict(X), color = "blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary (in millions $)")
plt.show()
```

## Visualising the Polynomial Regression results

```
# Visualizing the Polynomial Regression Data
# Copied from Linear Regression results
plt.scatter(X, y, color = "red")
# Modified predict to adapt polynomial regression model
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = "blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary (in millions $)")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
# Arrange min and max values of X to reshape line
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

# Copied from above
plt.scatter(X, y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = "blue")
plt.title("Truth or Bluff (Revised Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary (in millions $)")
plt.show()
```

## Predicting a new result with Linear Regression

```
# Prediction based on estimate on Linear Regression (Less Accurate)
lin_reg.predict([[6.5]])
```

## Predicting a new result with Polynomial Regression

```
# Prediction based on estimate on Polynomial Regression (More Accurate)
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
